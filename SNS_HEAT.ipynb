```python
# import 파트
import json
import os
import pandas as pd
from apyori import apriori
```

```python
json_path = '/content/drive/MyDrive/Colab Notebooks/test'
json_files = [pos_json for pos_json in os.listdir(json_path) if pos_json.endswith('.json')]
```

```python
extracted_data = []
for json_file in json_files:
  with open(os.path.join(json_path, json_file), 'r') as file:
    data = json.load(file)
    for item in data['info']:
      extracted_item = []
      extracted_item.append(item['id'])
      extracted_item.append(item['mediatype'])
      extracted_item.append(item['medianame'])
      extracted_item.append(item['category'])
      extracted_item.append(item['date'])
      extracted_item.append(item['annotations']['subject'])
      extracted_item.append(item['annotations']['speaker_type'])
      extracted_item.append(item['annotations']['lines'][0]['speaker']['sex'])
      extracted_item.append(item['annotations']['lines'][0]['speaker']['age'])
      extracted_data.append(extracted_item)

for item in extracted_data:
   print(item)
```

```python
df = pd.DataFrame(extracted_data, columns=['ID', 'mediatype', 'medianame', 'category', 'date', 'subject', 'speaker_type', 'sex', 'age'])
df['Data'] = df[['mediatype', 'medianame','subject', 'speaker_type', 'sex', 'age']].apply(lambda row: ', '.join(row.values.astype(str)), axis=1)
df = df.drop(['mediatype', 'medianame', 'category', 'subject', 'speaker_type', 'sex', 'age'], axis=1)
df.head()
```

```python
data = df.drop(['ID','date'], axis = 1)
data.head()
```

```python
records = []
for i in range(data.shape[0]):
    records.append([str(data.values[i,j]).split(',') for j in range(data.shape[1])])

trx = [[] for trx in range(len(records))]
for i in range(len(records)):
    for j in records[i][0]:
        trx[i].append(j)
trx
```

```python
association_rules = apriori(trx, min_support=0.1, min_confidence=0.80,min_lift=1)
association_results = association_rules

pd.set_option('max_colwidth', 1000)
Result=pd.DataFrame(columns=['Rule','Support','Confidence'])
for item in association_results:
    pair = item[2]
    for i in pair:
        items = str([x for x in i[0]])
        if i[3]!=1:
            new_row = pd.DataFrame({
                'Rule':[str([x for x in i[0]])+ " -> " +str([x for x in i[1]])],
                'Support':[str(round(item[1]*100,2))+'%'],
                'Confidence':[str(round(i[2] *100,2))+'%']
                })
            Result = pd.concat([Result, new_row], ignore_index=True)
Result
```
